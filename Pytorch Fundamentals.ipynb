```python
# Original tensor with multiple dimensions
tensor_3d = torch.rand(2, 3, 4)  # 2 × 3 × 4 = 24 elements total

# Valid reshapes (all have 24 elements):
tensor_3d, tensor_3d.reshape(24),tensor_3d.reshape(6, 4), tensor_3d.reshape(4, 1, 6),tensor_3d.reshape(1, 24, 1)     # 3D: 1 × 24 × 1 = 24 ✓

```

```python
tensor_3d.ndim, tensor_3d.reshape(24).ndim, tensor_3d.reshape(6, 4).ndim, tensor_3d.reshape(4, 2, 3).ndim, tensor_3d.reshape(1, 24, 1).ndim
```

```python
a = torch.tensor([[1, 2], [3, 4]])  # shape: (2, 2)

torch.stack([a, a])      # shape: (2, 2, 2) - new dimension added
```

```python
torch.vstack([a, a])     # shape: (4, 2) - stacked vertically
```

```python
torch.hstack([a, a])     # shape: (2, 4) - stacked horizontally
```
